```python
#loading necessary libraries
from randomuser import RandomUser
import pandas as pd
```

```python
#creating a random user object r
r=RandomUser()
```

```python
#using generate users() function we get a list of random users
list=r.generate_users(15)
list
```

```python
#generating full name using get_full_name() function from the get methods
name=r.get_full_name()
```

```python
#for loop for getting 15 users with their full name and email
for user in list:
    print(user.get_full_name()," ",user.get_email())
```

```python
#generating their photos
for user in list:
    print(user.get_picture())
```

```python
for user in list:
    print(user.get_password())
```

```python
#Generating a table with information abouut the users 
def get_users():
    users=[]
    for user in RandomUser.generate_users(10):
      users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),"Email":user.get_email(), "DOB":user.get_dob(),"Picture":user.get_picture()})
    return pd.DataFrame(users)
```

#### FRUITVICE API

```python
#using requests library
import requests
import json
```

```python
#we will obtain the fruitvice API using requests.get(url)function
data=requests.get("https://fruityvice.com/api/fruit/all")
```

```python
#retrive results using json.loads
results=json.loads(data.text)
```

```python
#converting jason data to panda dataframe
pd.DataFrame(results)
```

```python
#The result is in a nested json format.
#The 'nutrition' column contains multiple subcolumns, so the data needs to be 'flattened' or normalized.
df2=pd.json_normalize(results)
```

```python
#trying to extract information
apple=df2.loc[df2["name"] == 'Apple']
(apple.iloc[0]['family']),(apple.iloc[0]['genus'])
```
